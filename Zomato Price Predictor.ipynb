```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
data
```

```python
data.columns #what are the thinks that will affect the average cost for two
```

```python
#deleting unwanted columns permanently
data.drop(columns=['Restaurant ID','Is delivering now','Switch to order menu','Rating color','Rating text'],inplace=True)
```

```python
data.columns
```

```python
data=data[data['Country Code']==1] #we are masking because maximum datas are of country code =1
data
```

```python
data.shape
```

```python
data=data[data['City'].isin(['New Delhi','Gurgaon','Noida','Faridabad'])]
#again masking cause maximum information we are getting from these 4 cities
data
```

```python
data.columns
```

```python
data.drop(columns=['Restaurant Name','Country Code','Address','Locality','Locality Verbose','Currency'],inplace=True)
```

```python
data['Has Table booking'].replace({'Yes':1,'No':0},inplace=True)
```

```python
data['Has Online delivery'].replace({'Yes':1,'No':0},inplace=True)
```

```python
data.corr()['Average Cost for two']
```

```python
data.drop(columns=['Has Online delivery','Price range'],inplace=True)
```

```python
data.columns
```

```python
data.info()
```

```python
data
```

```python
data.head()
```

```python
a=data.groupby('Cuisines')['Average Cost for two'].mean().sort_values(ascending=False).reset_index() #creating a new data frame with the required steps
```

```python
data=data.merge(a,on='Cuisines')
```

```python
data.drop(columns=['Cuisines'],inplace=True)
```

```python
data.head()
```

```python
data.rename(columns={'Average Cost for two_x':'cost','Average Cost for two_y':'Cuisines'},inplace=True)
```

```python
data
```

```python
data.corr()['cost']
```

```python
y=data.iloc[:,3].values
```

```python
y
```

```python
y.shape
```

```python
x=data.drop(columns=['cost']).iloc[:,:].values #not a permanent step
```

```python
x #ek numpy array
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()  
```

```python
x[:,0]=encoder.fit_transform(x[:,0])
```

```python
from sklearn.preprocessing import OneHotEncoder
oencoder=OneHotEncoder(categorical_features=[0])
```

```python
x=oencoder.fit_transform(x).toarray()
```

```python
x.shape #WE can see columns added. with one column removed and 4 added
```

```python
x[:,0:4]
```

```python
x=x[:,1:]
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
from sklearn.linear_model import LinearRegression
regressor1=LinearRegression()
```

```python
regressor1.fit(x_train,y_train)
```

```python
y_pred=regressor1.predict(x_test)
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor2=DecisionTreeRegressor()
```

```python
regressor2.fit(x_train,y_train)
```

```python
y_pred2=regressor2.predict(x_test)
```

```python
r2_score(y_test,y_pred2)
```

```python
regressor1.coef_
```

```python
regressor1.intercept_
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor3=RandomForestRegressor(n_estimators=300)
```

```python
regressor3.fit(x_train,y_train)
```

```python
y_pred3=regressor3.predict(x_test)
```

```python
r2_score(y_test,y_pred3)
```

```python
from sklearn.svm import SVR
regressor4=SVR()
```

```python
regressor4.fit(x_train,y_train)
```

```python
y_pred4=regressor4.predict(x_test)
```

```python
r2_score(y_test,y_pred4)
```
