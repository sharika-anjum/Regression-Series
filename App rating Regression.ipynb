# now we'll work with all the words present under the category columns.we will count the probability of each words and store there values in the suitable columns

```python
category_words  = []
for i in data['Category']:
    blob=i.split("_")
    category_words +=blob
```

```python
category_words
```

```python
for i in range(len(category_words)):
    if not category_words[i].isalpha():
            category_words[i]=""
```

```python
count = Counter(category_words)
count
```

```python
for i in count:
    print(i)
```

```python
len (count)
```

```python
del count['AND']
```

```python
len(count)
```

```python
count
```

```python
count_words = []
for i in count:
    count_words.append(i)
```

```python
len(count)
```

```python
category_new=[]
for i in data['Category']:
    blob=i.split("_")
    data1=[]
    for j in count:
        data1.append(blob.count(j[0]))
    category_new.append(data1)
```

```python
data.shape
```

```python
len(category_new)
```

```python
new_column=np.array(category_new)
for i in range(len(count)):
    data[count_words[i]] = new_column[:,i]
```
